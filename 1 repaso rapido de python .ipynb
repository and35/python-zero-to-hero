````python
# por ejemplo si queremos crear una claculadora podmos frear una funcion para cada operacion para organizar y simplifcar 
def multiplicacion(a, b):
    """ 
aqui puedes colocar la documentacion
    """
    c = a * b
    print("multiplicaste %d * %d:  "%(a,b))
    return c 
resultado = multiplicacion(2, 3) # el resultado lo guardamo en las variables globales 
print("resultado: " + str(resultado))
````

# creando calculadora de IMC
## 1: buscar informacion para saber como plantear el problema 
Kilogramos y metros (o centímetros)	Fórmula: peso (kg) / [estatura (m)]^2  

Con el sistema métrico, la fórmula para el IMC es el peso en kilogramos dividido por la estatura en metros cuadrados. Debido a que la estatura por lo general se mide en centímetros, divida la estatura en centímetros por 100 para obtener la estatura en metros.  

Ejemplo: Peso = 68 kg, Estatura = 165 cm (1.65 m)
Cálculo: 68 ÷ (1.65)2 = 24.98
## 2 resolver el problema en modulos usando funciones 
esto nos permitira organizar el codigo, usar la mecanica de resolver problemas de "divide y venceras" ademas de ir probando modulo por modulo para ver si algo falla en la ultima celda vemos como realizamos eso con el modulo "evaluar imc"
## 3 crear el modulo 
usa todo lo aprendido 
## 4 une todas las funciones para que comience a funcionar 

````python
# ejercicio para calcular el IMC
print("-*" * 50) 
print("calculadora de indice de masa corporal ") 
print("-*" * 50)
def calculadora_imc(kg, m):
    imc = (kg) / (m)**2
    return imc

def evaluar_imc(imc):
    print(" tienes un IMC = " + str(imc))
    if imc <= 18.5:
        print("estas Bajo de peso")
    elif imc > 18.5 and imc <= 25: # usamos operadores logicos para definir un rango tambien se podria usar range(18.5, 25, 0.01)
        print("estas Normal en peso")
    elif imc > 25 and imc <= 29.99:
        print("tienes sobrepeso")
    else:
        print("tienes obesidad")

def inicio(): 
    # reciclamos codigo que usamos antes para validar si lo ingresado es un numero
    while True:
        try:
            num1 = input("ingresa tu peso en kg: ")
            num2 = input("ingresa estatura en mts: ")
            num1 = float(num1)  
            num2 = float(num2)
            return num1, num2 #asi se returanan 2 numeros 
            break
        except:
            print("no ingreso un numero, intentelo de nuevo. ")
            continue #si no funciona volvera al inicio 

kg, mts = inicio()
print("**" * 50)
evaluar_imc(calculadora_imc(kg, mts))
````
